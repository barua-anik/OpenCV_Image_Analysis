```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('pillow.jpg', 0)
img2 = cv2.imread('pillows.jpg', 0)

# Detector for similariries or ORB constructor

# ORB detector
orb = cv2.ORB_create()

# Key point and descriptor 
kp1, des1 = orb.detectAndCompute(img1, None)
kp2, des2 = orb.detectAndCompute(img2, None)

# Find out the Key point and descriptor using BF matcher 
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)
matches = bf.match(des1, des2)
# Creating matches for descriptors and sorting based on the distance
matches = sorted(matches, key=lambda x:x.distance)

# So, the final image shows best 20 matches
img3 = cv2.drawMatches(img1, kp1, img2, kp2, matches[:20], None, flags = 2)

plt.figure(figsize=(12,9))
plt.imshow(img3)
plt.show()
```
