```python
"""
This a simple program to load, manipulatea and
save images using OpenCV and matplotlib
"""
```

```python
# Importing libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the image using Opencv

img = cv2.imread('anik.jpg',cv2.IMREAD_UNCHANGED)

# Showing the image

resized = cv2.resize(img, (500,300), interpolation = cv2.INTER_AREA)
cv2.imshow('image',resized)  

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Showing the image in matplotlib

plt.imshow(img, cmap = 'winter', interpolation = 'mitchell')
```

```python
# Saving the image file

cv2.imwrite('anik_openCV_pic.jpg',img)
```
